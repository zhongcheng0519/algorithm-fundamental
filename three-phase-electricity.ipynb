# 三相电

## 工频数据，相电压

```python
import numpy as np
import matplotlib.pyplot as plt

t0 = 0
t1 = 0.1
nsamples = 500
t = np.linspace(t0, t1, nsamples)
dt = (t1-t0)/nsamples
f = 50     # 频率为50Hz
w = 2*np.pi*f

U = 220
Up = 220*np.sqrt(2)
AN = Up*np.sin(w*t)
BN = Up*np.sin(w*t-2/3*np.pi)
CN = Up*np.sin(w*t-4/3*np.pi)

fig = plt.figure(figsize= (18, 6))

ax = fig.add_subplot(1, 1, 1)
ax.plot(t, AN, 'r-', label='A')
ax.plot(t, BN, 'g-', label='B')
ax.plot(t, CN, 'b-', label='C')
plt.legend()
plt.show()
```

## 线电压

```python
U_ab = AN - BN
U_bc = BN - CN
U_ca = CN - AN

fig = plt.figure(figsize= (18, 6))

ax = fig.add_subplot(1, 1, 1)
ax.plot(t, U_ab, 'r-', label='ab')
ax.plot(t, U_bc, 'g-', label='bc')
ax.plot(t, U_ca, 'b-', label='ca')
plt.legend()
plt.show()
```
